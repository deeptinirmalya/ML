```python
import json
```

```python
def load_dta(filename):
  with open(filename,"r") as f:
    data = json.load(f)
    return data
```

```python
data = load_dta("amazon_data.json")
print(type(data))
```

```python
# cleane and structure the data
def clean_data(data):
  text_to_number = {"one":1, "two":2, "three":3, "four":4, "five":5}
  cleaned_data = []
  unique_user = set()

  for user in data:
    raw_rating = str(user["rating"]).strip().lower()
    if(raw_rating in text_to_number):
        raw_rating = text_to_number[raw_rating]
    user["rating"] = raw_rating

# handel missing value
    raw_age = user.get("age")
    if raw_age == None:
      user["age"] = None
# duplicate data handel
# in this i acn just take out the unique id so the duplicate are handeled
   # perform de duplication
    if(user["name"].strip() in unique_user):
      continue
    unique_user.add(user["name"])
    cleaned_data.append(user)
  return cleaned_data
```

```python
data = clean_data(data)
```

```python
# get the meaning ful insight
def get_insights(data):
  # avrage rating
  total_rating = 0
  for user in data:
    total_rating  += float(user["rating"])
  print(f"avrage rating= {total_rating/len(data)}")

  # % of user with poor rating
  poor_rating = 0
  for user in data:
    if float(user["rating"]) <= 3:
      poor_rating += 1
  print(f"poor rating % = {poor_rating/len(data)*100}%")

```

```python
get_insights(data)
```

```python
# recomendation feature
def get_recomendation(data):

  recomendation = []

  for user in data:
      current_recom = {}
      current_recom["name"]  = user["name"]
      if float(user["rating"]) >= 4:
        current_recom["brand"] = "apple"
      else:
        current_recom["brand"] = "sumsung"

      recomendation.append(current_recom)

  return recomendation


```

```python
get_recomendation(data)
```
